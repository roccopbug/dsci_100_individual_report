```r
library(tidyverse)
```

```r
players <- read_csv("data/players.csv")

players <- players|>
            mutate(experience = as_factor(experience),
                  gender = as_factor(gender),
                  subscribe = as_factor(subscribe))
players
```

- Our dataset has 196 rows, meaning we have 196 observations/players with data. The data-set is showing characteristics of a player. There are 7 variables, 
- experience: a class(changed to factor from character in case we need it as a class to predict), how experienced is a player in Minecraft? 5 options, a class we could predict.
- subscribe: a boolean True (144) or False (52), whether the player is subscribed to a game-related newsletter. Another class we can predict, will a player subscribe to a newsletter based on age? gender?...
- hashedEmail: participants encrypted email, not needed
- played_hours: hours a participant played on the Minecraft server (1 decimal)
- name: participant name
- gender: participant gender, 7 options
- Age: participant age
- The data is tidy, each row is 1 observation/player, each column is 1 variable, and values don't spread across columns

```r
players_means <- summarize(players, 
                            mean_age = mean(Age, na.rm = TRUE),
                            mean_hours = mean(played_hours, na.rm = TRUE))|>
                                mutate(mean_age = round(mean_age, 2), mean_hours = round(mean_hours, 2))
players_means

subscribed_counts <- players|>
                group_by(subscribe)|>
                    summarize(count = n())
subscribed_counts

experience_counts <- players|>
                group_by(experience)|>
                    summarize(count = n())
experience_counts

gender_counts <- players|>
            group_by(gender)|>
            summarize(count = n())
gender_counts
```


- mean_age: 21.14 years, some data removed due to missing values
- mean_hours: 5.85 hours
- Counts done to get an overall sense of the data, will do plots later in the notebook

**Question:** can played_hours and age predict if a player is subscribed to a gaming newsletter? This is a classification problem, so we'll be using k-nearest neighbors to see if we can make a model that can predict subscription status based on hours played and age. I already turned subscribtion into a factor so other than that, no wrangling is needed.

**Exploratory vizualizations:**

```r
experience_prop <- players|>
            group_by(experience)|>
            summarize(prop = round(n()/nrow(players),2))
experience_prop

experience_plot <- ggplot(players, aes(x = experience, fill = experience)) + 
            geom_bar() + 
            labs(x = "Player Experience level", y = "Count", fill = "Player Experience level") + 
            ggtitle("Counts of players experience levels")
experience_plot
```

Most players are more casual

```r
gender_plot <- ggplot(players, aes(x = gender, fill = gender)) + 
        geom_bar()+
        theme(text = element_text(size = 10))+
        labs(x = "Player Gender", y = "Count", fill = "Player Gender") + 
        ggtitle("Counts of players Genders")
gender_plot

gender_prop <- players|>
        group_by(gender)|>
        summarize(prop = round(n()/nrow(players),2))
gender_prop
```

Males are the most frequent players.

```r
age_hours_plot <- ggplot(players, aes(x = Age, y = played_hours + 1, colour = subscribe)) + #avoid log(0)
                    geom_point()+
                    scale_y_log10()+
                    labs(x = "Age", y = "total hours played (scaled by log10 + 1)", colour = "Subscribed to\nGame-related\nNewsletter?")+
                    ggtitle("Hours played based on age")
age_hours_plot

subscribe_prop <- players|>
        group_by(subscribe)|>
        summarize(prop = round(n()/nrow(players),2))
subscribe_prop
```

The age hours plot helps show whether age affects playtime. I thought hours would decrease with age, but we see 15-25 year olds vary a lot. We’re also lacking enough older data to see anything. I scaled hours played with log + 1 to avoid log0/negative values. From the coloured points, we can see with higher ages, or less hours we generally see more red points compared to younger, more frequent players.

- knn is appropriate since we're trying to predict a categorical variable, so we need to use classification rather than regression.

- We're gonna assume that hours and age can be used to predict experience levels. Whether that's the case we'll figure out by doing the knn classification

- knn gets slow if we have very large datasets as it has to calculate the distance between many points. This shouldn't be an issue with the players dataset, but in the future if we get more data it could turn into a problem. Another potential issue, most of the players are in the TRUE class rather than it being a even split. This could bias our model, leading to alot of TRUE predictions simply because there are more TRUE points.

- We'll use cross validation to figure out what is the optimal k value to give us the most accurate predicting model. 

- We'll split the data into training (70%) and testing (30%). We will also scale our predictors at the start so that hours played doesn't hold more power over age. We split the data at the start, we'll use the training data to fit the model, then at the end we use it on the testing data. We'll use cross validation to figure out the optimal k value for the amount of neighbors that give us the most accurate predictions.
